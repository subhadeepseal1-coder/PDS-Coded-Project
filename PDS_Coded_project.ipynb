**E1." Does Men tend to prefer SUVs more compared to women?"**

Based on the plot, we can see the gender distribution for each car type.
By examining the plot , we can determine if the number of males who prefer SUVs are considerably higher than females.


```python
#Create a bar plot to compare gender distribution for each car type
sns.countplot(df, x="Make", hue="Gender", color="Orange")
plt.title("Gender Distribution for each car type")
plt.xlabel("Car Type ")
plt.ylabel("Count")
plt.legend(title="Gender", loc="upper right", labels=["Male", "Female"])
plt.show()
```

**ANSWER**

From the above bar plot it is evident that females prefer SUV cars more than males.Hence the answer is **"No"**

**E2."Does A salaried person is more likely to buy a Sedan?"**

By examining the plot, we will be able to understand whether a salaried person(likely to have a personal loan) prefers Sedans more than any other car types.

```python
#Create a bar plot top compare the distribution of "Personal_loan" for each car type
sns.countplot(df, x="Make", hue="Personal_loan")
plt.title(" Personal loan Distribution for each car type")
plt.xlabel("Car Type")
plt.ylabel("Count")
plt.legend(title="Personal loan", loc="upper right", labels=["No", "Yes"])
plt.show()
```

**ANSWER**

From the above bar plot it is evident that a salaried person is more likely to buy a Sedan.Hence the answer is **"Yes"**.

**E3.What evidence or data supports Sheldon Cooper's claim that a salaried male is an easier target for a SUV sale over a Sedan sale?**

By examining this plot, we can find if the salaried males show a higher preference for SUVs compared to Sedans.
Justification:If Sheldon's claim is true , we should observe a higher number of salaried males choosing SUVs over Sedans compared to other groups(salaried females and non-salaried males and females).

**E4.How does the the amount spent on purchasing automobiles vary by gender?**

**Justification**:
If we observe a significant difference between males and females , the business can utilize this information to tailor marketing strategies and promotions for each gender.For example, if males tend to spend more, the business can offer  premium or high end car models , while for females, they can focus on providing options that align with their preferences and budget.

```python
#Explore the total purchase on automobiles by gender using a bar plot
sns.barplot(df, x="Gender",y="Price", hue="Gender", errcolor="black", capsize=0.1)
plt.title("Total Purchase on automobiles by Gender ")
plt.xlabel("Gender")
plt.ylabel("Total Purchase")
plt.show()
```

**E5.How much money was spent on purchasing automobiles by individuals who took a personal loan? **

Justification:If we observe a difference in the distribution of spending between customers with and without personal loans, the business can utilize this information to create targeted offers.For instance, customers with personal loans may have different financial considerations , so the business can offer financing options or deals that cater to their specific needs.On the other hand , customers without personal loans may prefer one-time payment options or incentives for immediate purchase.

```python
#Creating a box plot to compare the distribution of spending for customers with and without personal loans
sns.boxplot(df, x="Personal_loan",y="Price", hue="Personal_loan")
plt.title("Total spending on Automobiles by Personal loan")
plt.xlabel("Personal_loan")
plt.ylabel("Total Spending")
plt.xticks(ticks=[0, 1],labels=["No","Yes"])
plt.show()
```
